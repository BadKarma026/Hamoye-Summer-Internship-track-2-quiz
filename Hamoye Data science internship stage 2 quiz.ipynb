# QUESTION 12:
From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two d.p.?

```python
sample_df = df[['T2', 'T6']].sample(100, random_state=42)
sns.regplot(x="T2", y="T6",data=sample_df)
plt.xlabel('Temperature in the Living Room')
plt.ylabel('Temperature outside the Building')
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

scaler = MinMaxScaler()

normalised_df_1 = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

feature_variable_1 = normalised_df_1['T2']
target_variable_1 = normalised_df_1['T6']

x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(feature_variable_1, target_variable_1, test_size=0.30, random_state=42)

x_train_1 = x_train_1.values.reshape(-1, 1)
y_train_1 = y_train_1.values.reshape(-1, 1)
x_test_1 = x_test_1.values.reshape(-1, 1)
y_test_1 = y_test_1.values.reshape(-1, 1)

from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

linear_model.fit(x_train_1, y_train_1)

predicted_values_1 = linear_model.predict(x_test_1)

from sklearn.metrics import r2_score
r2_score = r2_score(y_test_1, predicted_values_1)
round(r2_score, 2)
```

# ANSWER 12:

# QUESTION 17:
Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?

```python
def get_weights_df(model, feat, col_name):

    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

linear_model_weights = get_weights_df(linear_model, x_train_2, 'Linear_Model_Weight')
linear_model_weights['Linear_Model_Weight'].min()
```

```python
linear_model_weights.query("Linear_Model_Weight == -0.4566979483385004")["Features"]
```

### Feature with minimum weight is "RH_2"

```python
linear_model_weights['Linear_Model_Weight'].max()
```

```python
linear_model_weights.query("Linear_Model_Weight == 0.5535465998386401")["Features"]
```

### Feature with maximum weight is "RH_1"

# ANSWER 17:
Features having the lowest and highest weights respectively are (RH_2, RH_1)

# QUESTION 18:
Train a ridge regression model with an alpha value of 0.4. Is there any change to the root mean squared error (RMSE) when evaluated on the test set?

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

scaler = MinMaxScaler()

normalised_df_3 = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

feature_variable_3_df = normalised_df_3.drop(columns='Appliances')
target_variable_3 = normalised_df_3['Appliances']

x_train_3, x_test_3, y_train_3, y_test_3 = train_test_split(feature_variable_3_df, target_variable_3, test_size=0.30, random_state=42)

from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train_3, y_train_3)
```

```python
predicted_values_3 = ridge_reg.predict(x_test_3)

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test_3, predicted_values_3))
round(rmse, 9)
```

# QUESTION 19:
Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. How many of the features have non-zero feature weights?

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

scaler = MinMaxScaler()

normalised_df_4 = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

feature_variable_4_df = normalised_df_4.drop(columns='Appliances')
target_variable_4 = normalised_df_4['Appliances']

x_train_4, x_test_4, y_train_4, y_test_4 = train_test_split(feature_variable_4_df, target_variable_4, test_size=0.30, random_state=42)

from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train_4, y_train_4)

predicted_values_4 = lasso_reg.predict(x_test_4)

lasso_weights_df = get_weights_df(lasso_reg, x_train_4, 'Lasso_weight')
lasso_weights_df.groupby('Lasso_weight').count()
```
